```python

import pandas as pd
import numpy as np
print(df.isnull().sum())
print(df.dtypes)
df = df.drop_duplicates()
df = df.dropna(subset=['answer'])
```

```python
import numpy as np

def assign_category(row):
    subject = str(row['subject']).lower() if isinstance(row['subject'], str) else str(row['subject'])
    body = str(row['body']).lower() if isinstance(row['body'], str) else str(row['body'])
    tag_1 = str(row['tag_1']).lower() if isinstance(row['tag_1'], str) else str(row['tag_1'])
    tag_4 = str(row['tag_4']).lower() if isinstance(row['tag_4'], str) else str(row['tag_4'])
    queue = str(row['queue']).lower() if isinstance(row['queue'], str) else str(row['queue'])

    if "billing" in subject or "billing" in body:
        return "Billing Inquiries"
    elif "critical" in subject or "urgent" in tag_1:
        return "Urgent Issues"
    elif "sales" in subject or "sales" in tag_4:
        return "Sales Inquiries"
    elif "support" in queue:
        return "Technical Support"
    else:
        return "General Inquiry"

df['category'] = df.apply(assign_category, axis=1)
```

```python
df.head(2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['combined_text'] = df['subject'].fillna('') + ' ' + df['body'].fillna('')
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['answer'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head()
```

```python

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(tfidf_df)
principal_components = pca.transform(tfidf_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

pca_df.head()
```

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative.fit(tfidf_df)
```

```python

df.loc[:, 'label'] = agglomerative.labels_
df.loc[df['id'].isin(df['id']), 'label'] = df['label'].values
answer= df.loc[df["label"].eq(2), :]["answer"].sample(frac=0.1)
```
